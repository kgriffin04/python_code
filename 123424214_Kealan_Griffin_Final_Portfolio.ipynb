# Part 2 - Plots and analysis

I want to understand the defensive performance of goalkeepers throughout the UCL 21/22 season regarding their **saves** and **cleansheets** produced during the tournamnet, hoping to identify trends in goalkeeping effectiveness and consistency over the course of the competition.

````python
saves_per_match = df_goalkeeping.groupby(['match_played', 'position'])['saved'].sum().unstack()
cleansheets_per_match = df_goalkeeping.groupby(['match_played', 'position'])['cleansheets'].sum().unstack()

plt.figure(figsize= (10,6))

# Plot for Saves
for position in saves_per_match.columns:
  plt.plot(saves_per_match.index, saves_per_match[position], label=f'{position} - Saves', linestyle = '-', color = 'red')

#Plot for Cleansheets
for position in cleansheets_per_match.columns:
  plt.plot(cleansheets_per_match.index, cleansheets_per_match[position], label=f'{position} - Cleansheets', linestyle = '--', color = 'green')

plt.title('Saves and Cleansheets by Goalkeeprs per Match - UCL 21/22')
plt.xlabel('Match Number')
plt.ylabel('Count')
plt.legend(title= 'Legend')
plt.grid(True)
plt.show()

df_goalkeeping[df_goalkeeping['player_name'] == 'Courtois']
````

Data displayed above showcases a numerous amount of insights:

*   **Pressures of the Group Stages:**
As seen around Match 6 both saves and cleansheets peak here, where qualification for the knockout rounds intensifies. Goalkeepers often face increased pressure here - not just from elite forwards, but also from determined strikers of 'smaller' clubs eager to make their mark. These forwards may lack the clinical finishing of top-tier attackers, leading to more shots on target but fewer goals. In response, experienced and composed goalkeepers rise to the occasion, reinforcing their team's progression hopes with solid defensive displays  
*   **Variability in Consistency:**
While a general relationship exists between saves and cleansheets, the pattern throughout the competition is far from uniform. Performance trends are staggered, with dips and spikes that reflect the unpredictable nature of knockout-football. These fluctuations highlight stand-out goalkeeping performances duriing pivotal moments, suggesting that key matches often bring out the best in certain players, even when under immense pressure   
*   **The 'Brick Wall' - Thibaut Courtois:**
Notably at Match 13 we can see a very strong spike in performance, this can be credited to a world-class performance from Thibaut Courtois throughout the competiton with 61 saves & 5 cleansheets, the Belgian keeping a cleansheet against Liverpool in the final making a total of 9 saves in this nail-bitting match alone, significantly aiding Real Madrid's victory over the Reds

## Part 2 - Univariate Analysis and Box Plot
I want to investigate the scoring patterns of players in different positions based on foot preference, maintaing focus on the regular starters by filtering players who have at least played 8 or more matches across the UCL 21/22 season, drawing a comparison between right and left footed goals from **'Forward'**, **'Midfielder'**, and **'Defender'**, seeking to identify **positional trends in finishing** as well as **foot dominance**

````python
# Using .melt here from the Pandas library complements the use of Seaborn in my plot
# rather than having to plot separate plots for my analysis


df_eligible = df_goals[df_goals['match_played'] >= 8] #  I have chosen to filter players that have played at least 8 or more matches to distinguish regular starting players in the competition

df_reshaped = df_eligible.melt(id_vars = ['position'],
                               value_vars = ['right_foot', 'left_foot'], # Makes 'right_foot' and 'left_foot' = 'Foot' (new column)
                               var_name = 'Foot',
                               value_name = 'Goals')

sns.boxplot(x = 'Foot',
            y = 'Goals',
            hue = 'position',
            data = df_reshaped,
            palette = {'Forward': 'blue', 'Midfielder': 'green', 'Defender': 'orange'})

plt.title('Distribution of Right vs Left Footed Goals UCL 21/22 Season (8+ Matches Played)')
plt.xlabel('Foot Used')
plt.ylabel('Goals')
plt.legend(title = 'Position')
plt.grid(True)
plt.show()

print('Right-Footed Goal Stats')
print(df_eligible.groupby('position')['right_foot'].describe())
print('\n')
print('Left-Footed Goal Stats')
print(df_eligible.groupby('position')['left_foot'].describe())
````

The analysis confirms expected trends of **'Forward'**- the primary goal scorers, **'Midfielder'**- more balanced, and **'Defender'**- scoring infrequently.
The mean of the right-footed goals for **forwards(1.97)** is significantly higher than for **midfielders(0.88)** and **defenders(0.64)**
The **standard deviation among forwards is higher** suggesting that goal-scoring ability is **more unevenly distributed**- Considering the competitiveness of the Champions League a few elite forwards score heavily, while others contribute less. Overall the right foot is the dominant scoring foot across all positions.


## Part 3 - Univariate Analysis and Box Plot
I want to investigate the **distribution of different goal types of the top 5 scorers** within the competition. Examining:

*   right_footed
*   left_footed
*   headers
*   penalties
*   other

...as well as aiming to underscore their individual approach to their role as a goalscorer

````python
df_goals['total_goals'] = df_goals[['right_foot','left_foot','headers', 'penalties', 'others']].sum(axis = 1)

top_5 = df_goals.sort_values(by = 'total_goals', ascending = False).head(5).set_index('player_name')

goal_type = ['right_foot', 'left_foot', 'headers', 'penalties', 'others']

top_5[goal_type].plot(kind = 'box',
                      title = "Top 5 Scorers' Goal Type Distribution UCL 21/22",
                      grid = True,
                      figsize = (8,6))

for player in top_5.index:
  plt.scatter(range(1, len(goal_type) +1), top_5.loc[player, goal_type], s = 100, alpha = 0.4, label = player) # Plots a Scatter Plot inside a Box Plot- being able to highlight individual players
                                                                                                               # Creates a value for the player on the x-axis, plotted aganist sum of goal types
plt.legend(title = 'Players')
plt.xlabel('Goal Type')
plt.ylabel('Goals Scored')
plt.show()

print("Top 5 Scorers' Goals")
print(top_5[goal_type].T) # Transposes players and goal types
print('\n')
print("Top 5 Scorers' Goal Stats")
print(top_5[goal_type].describe())
````

Findings from the results show some extensive insights into the distribution across the different goal types of the top 5 scorers:

*   Right-footed goals dominate- The most common scoring method among the top 5 with **Karim Benzema (11)** and **Robert Lewandowski (8)** leading- **mean right-footed goals (4.6)** suggesting a strong preference for right-footed shooting
*   **Left-footed goals (4 avg, std: 2.55)** showing that **left-footed strikers contribute significantly more but vary in volume**- Mohamed Salah & Riyad Mahrez primarly left-footed finishers, with fewer right-footed goals than others
*   **Headers are balanced** with most players scoring 1-3 headed goals- no extreme outliers
*   **A low variance in penalties (std:10)**- consistency in opportunites for penalties
*   **Other goal types being the least frequent** category, with only **Haller** and **Lewandowski** scoring once each



---

Each player displays various methods of scoring, highlighting the individual approach each striker takes to fulfill their role as a goalscorer, Benzema & Lewandowski show a diverse scoring profile, Salah & Mahrez demonstrate the effectiveness of a dominant left-foot and Haller's headers indicate a strong aerial presence.



## Part 2 - Scatter Plot
I want to investigate a correlation between **'pass_attempted'** and **'pass_accuracy'**, seeing how pass attempts influence pass accuracy and whether players who attempt more passes tend to have higher accuracy.


````python
plt.figure(figsize = (8,6))
sns.regplot(x = df_passing['pass_attempted'], # Use a regression line to improve plot
            y = df_passing['pass_accuracy'],
            scatter_kws = {'alpha':0.6, 'color': 'orange'},
            line_kws = {'color':'blue'},
            ci = None)

plt.title('Pass Attempts vs Accuracy')
plt.xlabel('Passes Attempted')
plt.ylabel('Passing Accuracy')
plt.grid(True)
plt.show()
````

From the data displayed above, the plot suggests that **pass accuracy genrally improves as players attempt more passes**. Some other factors to consider:

*   Players with fewer pass attempts (left side of the plot) shows a **wider range of accuracy values**, some having very low accuracy and others reaching close to 100%
*  Players with more pass attempts (right side of the plot) tend to **cluster on the higher accuracy values (above 80% roughly)**, which indicates that **players who attempt more passes generally maintain efficent passing accuracy**
*  The dense cluster seen in the plot suggest **most players** fall in around **50-200 Pass Attempts & 70-90% Accuracy**

*  Notably one outlier close to the **500 marker of passes attempted**, had a **passing accuracy below 70%**, this player could been seen to have given the ball away a lot of the time, **underperforming from the rest**





# Part 3 - Area Plot

I want to understand how different player positions contributed to goal-scoring throughout 21/22 UCL season, grouping the data by **match_played** and **position**, plotted with the total goals scored by each position across matches utilizing an Area Plot
